```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from neuralforecast import NeuralForecast
from neuralforecast.models import TSMixerx
from neuralforecast.losses.pytorch import DistributionLoss

def append_data(filename, uf, start_date, end_date, val_id, var_name):
    df = pd.read_csv(filename, parse_dates=['date'])
    mask = (
        (df['uf'] == uf) &
        (df['date'] >= pd.to_datetime(start_date)) &
        (df['date'] <= pd.to_datetime(end_date)) &
        (df['val_set_id'] == val_id)
    )
    return df.loc[mask, var_name].reset_index(drop=True)

train_data = pd.read_csv('../data/processed/train_std_state_1.csv', parse_dates=['date'])
val_data   = pd.read_csv('../data/processed/val_std_state_1.csv',   parse_dates=['date'])

train_begin = train_data['date'].min().strftime('%Y-%m-%d')
train_end   = train_data['date'].max().strftime('%Y-%m-%d')
val_begin   = val_data['date'].min().strftime('%Y-%m-%d')
val_end     = val_data['date'].max().strftime('%Y-%m-%d')

date_rng_train = pd.date_range(train_begin, train_end, freq='W')
date_rng_val   = pd.date_range(val_begin,   val_end,   freq='W')

ufs      = train_data['uf'].unique()
n_states = len(ufs)
ncols    = 2
nrows    = math.ceil(n_states / ncols)
fig, axs = plt.subplots(nrows, ncols, figsize=(20, 5 * nrows),
                        sharex=True, sharey=True, constrained_layout=True)
axs = axs.flatten()

all_preds = []

for idx, uf in enumerate(ufs):
    tr = train_data[train_data['uf'] == uf].copy()
    va = val_data[  val_data['uf']   == uf].copy()

    pdo_data        = append_data('context_pdo.csv',          uf, val_begin, val_end,1,'pdo')
    temp_min_data   = append_data('context_temp_min.csv',     uf, val_begin, val_end,1,'temp_min')
    precip_min_data = append_data('context_precip_min.csv',   uf, val_begin, val_end,1,'precip_min')
    pressure_data   = append_data('context_pressure_min.csv', uf, val_begin, val_end,1,'pressure_min')

    tr['unique_id'] = 'regional_geocode'
    va['unique_id'] = 'regional_geocode'
    tr['ds'] = date_rng_train
    va['ds'] = date_rng_val
    tr.rename(columns={'cases':'y'}, inplace=True)
    va.rename(columns={'cases':'y'}, inplace=True)

    Y_train = tr.drop(columns=['date','epiweek','uf','year'])
    Y_test  = va.drop(columns=['date','epiweek','uf','year',
                               'temp_min','precip_min','pressure_min','pdo'])
    Y_test['pdo']          = pdo_data.values
    Y_test['temp_min']     = temp_min_data.values
    Y_test['precip_min']   = precip_min_data.values
    Y_test['pressure_min'] = pressure_data.values
    prediction_intervals=PredictionIntervals()
    h = len(Y_test)
    model = TSMixerx(
        h=h,
        input_size=10,
        n_series=1,
        hist_exog_list=['temp_min','precip_min','week_sin','week_cos','population','pdo','pressure_min'],
        futr_exog_list=['temp_min','precip_min','week_sin','week_cos','population','pdo','pressure_min'],
        n_block=4,
        ff_dim=24,
        dropout=0.1,
        revin=True,
        loss=MSE(),
        max_steps=250
    )

    fcst = NeuralForecast(models=[model], freq='W')
    fcst.fit(df=Y_train, prediction_intervals= prediction_intervals)
    forecasts = fcst.predict(futr_df=Y_test, level=[50,80,90,95])

    df_pred = pd.DataFrame({
        'uf':    uf,
        'date':  forecasts['ds'],
        'pred':  forecasts['TSMixerx'],
        'lo50':  forecasts['TSMixerx-lo-50'],  'hi50':  forecasts['TSMixerx-hi-50'],
        'lo80':  forecasts['TSMixerx-lo-80'],  'hi80':  forecasts['TSMixerx-hi-80'],
        'lo90':  forecasts['TSMixerx-lo-90'],  'hi90':  forecasts['TSMixerx-hi-90'],
        'lo95':  forecasts['TSMixerx-lo-95'],  'hi95':  forecasts['TSMixerx-hi-95'],
    })
    all_preds.append(df_pred)

    ax = axs[idx]
    ax.plot(forecasts['ds'], forecasts['TSMixerx'], color='blue', label='Forecast')
    ax.fill_between(
        forecasts['ds'],
        forecasts['TSMixerx-lo-90'],
        forecasts['TSMixerx-hi-90'],
        alpha=0.3,
        label='90% interval'
    )
    ax.plot(Y_test['ds'], Y_test['y'], color='red', label='Actual')
    ax.set_title(uf, fontsize=8)
    ax.legend(fontsize=6)
    ax.set_ylabel('Incidence', fontsize=7)

for j in range(n_states, len(axs)):
    axs[j].set_visible(False)

plt.show()

df_all_predictions_1 = pd.concat(all_preds, ignore_index=True)
```
